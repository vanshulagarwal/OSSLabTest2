```python
#Q2
import numpy as np
```

```python
def func(arr):
    maxstr=""
    minstr=""
    maxlen=0
    minlen=1000000
    longest=False
    for str in arr:
        if len(str)>5:
            longest=True
        if len(str)>maxlen:
            maxlen=len(str)
            maxstr=str
        if len(str)<minlen:
            minlen=len(str)
            minstr=str
    if longest:
        return maxstr
    else:
        return minstr

arr=np.array(["abcdefg","abc","abcdef","a"])
print(func(arr))
arr=np.array(["abcd","b","abc","ab","a"])
print(func(arr))
```

```python
#Q1
from sklearn.datasets import load_iris
import pandas as pd
```

```python
df=load_iris(as_frame=True)
```

```python
dt=df['data']
```

```python
#4.
from sklearn.model_selection import train_test_split

y=df['target']

X_train,X_test,Y_train,Y_test=train_test_split(dt, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()

ss.fit(X_train)
x_scaled_train=ss.transform(X_train)
```

```python
import matplotlib.pyplot as plt

plt.figure()

# plt.subplot(1,1,1)
plt.hist(X_train)
plt.hist(x_scaled_train)


plt.figure()
plt.boxplot(X_train)
plt.boxplot(x_scaled_train)
```
